```python
# Example data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]

# Create a figure with 4 subplots (2x2 grid)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Add plots to each subplot
for i, ax in enumerate(axs.flat):
    ax.plot(x, y)
    ax.set_title(f"Plot {i+1}: A longer title to show the difference")
    ax.set_xlabel("X-axis Label")
    ax.set_ylabel("Y-axis Label")

# Show the figure without tight layout
plt.show()

# Now let's use tight_layout() to adjust spacing
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Add plots to each subplot
for i, ax in enumerate(axs.flat):
    ax.plot(x, y)
    ax.set_title(f"Plot {i+1}: A longer title to show the difference")
    ax.set_xlabel("X-axis Label")
    ax.set_ylabel("Y-axis Label")

# Apply tight layout to adjust spacing
plt.tight_layout()

# Show the figure with tight layout
plt.show()
```

```python
# 3 in one
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2]) # left, bottom, width, height (range 0 to 1)
```

```python
ax1.plot(x,y)
ax1.set_xlabel('x')
ax1.set_ylabel('y')


ax2.plot(x,y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')

fig # Show figure object
```
