```python
# 2. Given the sample dataset [1, 135, 12, 18, 20, 2, 13, 17, 19, 36], calculate the mean, variance, and standard deviation manually. Compare your manual calculations with the Python results to ensure your understanding of these statistical measures. (4 marks)
#    (Offline calucator use is allowed)

dataset = [1,135,12,18,20,2,13,17,19,36]

#mean
sum = 0
for i in dataset:
    sum = sum + i
mean = sum/len(dataset)
print(mean)

#variance
sum = 0
for i in dataset:
    sum = sum + ((i**2) - mean**2)/len(dataset)
variance = sum
print(variance)

#standard deviation
std = math.sqrt(variance)
print(std)
```

```python
# 3. Given an input string s and an input integer n. Write a python code that appends to s the reverse of s in the first iteration, then appends to the new s(string) the original s in secode iteration, then again appends the reverse of s in the third iteration and so on tilll n iterations. (4 marks)
#     Example: Input string "ab", input n = 3   Answer: "abbaab" (ab + ba + ab) total 3 additions as n = 3, each addition is reverse previous addition

s = input("enter a string: ")
n = int(input("enter the no. of iterations: "))

j = -1
string= ""
for i in range(n):
    string = string + s[::-j]
    j = j * -1

print(string)
```
