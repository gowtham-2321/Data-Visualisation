```python
import seaborn as sns # pip install seaborn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Plotting KDE(Kernel Density Estimate)

# Sample data
data = [1, 2, 2, 3, 5]

sns.kdeplot(data, bw_adjust=0.5)
plt.show()
#KDE to estimate the PDF of a continuous random variable directly from the data
```

```python
sns.rugplot(data)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(20)
```

```python
#distribution plot
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
#main axes are bivariate.
#marginal axes are univariate
# What do you generaly infer from this?
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
'''
# Points to consider
1. Density
    a.Concentration of data in certain pockets e.g central tendency
    b.Potential Outliers
2. General trend of data
3. Correlation of any kind.
'''
```

```python
#Lets look at correlation of this data.
# Select only numeric columns
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])
print(numeric_tips.head())
```

```python
# Calculate the correlation matrix using only the numeric columns
correlation_matrix = numeric_tips.corr()
print(correlation_matrix)
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(8, 6))

# Create a heatmap using Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Show the plot
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# PLotting across categories
sns.pairplot(tips,hue='sex',palette='coolwarm')
```
